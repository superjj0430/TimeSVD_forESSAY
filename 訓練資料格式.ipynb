```python
import pandas as pd
```

```python
df = pd.read_csv('mid_user_convert.csv')
del df['Unnamed: 0']
del df['Author']
del df['Mid']
del df['Date']
print(df.info())
df
```

# 正規化timestamp to 0-100

```python
df["timestamp"]=((df["timestamp"]-df["timestamp"].min())/(df["timestamp"].max()-df["timestamp"].min()))*100
df
```

```python
# 換列順序
df = df[['user_convert', 'id_convert', 'Rating', 'timestamp']]
df
```

```python
# timestamp取整數
for i in range(16645):
    df['timestamp'][i] = round(df['timestamp'][i])
df
```

```python
df.astype({'timestamp': 'int64'}).dtypes
```

```python
d_dropany = df.dropna(axis=0, how='any')
d_dropany.isnull().any()
```
